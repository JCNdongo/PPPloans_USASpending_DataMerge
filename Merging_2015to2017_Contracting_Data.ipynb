### Merging all Contracting Data from 2015 to 2017 in one file ###

```python
#importing dependencies
import pandas as pd
```

```python
#define the directory path where each original file is located
directory_path = "F:\\FAU PhD\\DISSERTATION\\DATABASES\\DataTESTS\\2023.08.27_Test\\FinalDataMerge\\"
```

```python
#read each original csv file into a a Pandas DataFrame
df_2015 = pd.read_csv(directory_path + "All_2015to2016_Contracting.csv")
df_2017 = pd.read_csv(directory_path + "New_2017_Contracting.csv")
```

```python
#check that the 2015 data has correctly transferred from csv files into DataFrames. 
df_2015.head()
```

```python
#check that the 2015 data has correctly transferred from csv files into DataFrames. 
df_2015.shape
```

```python
#check that the 2016 data has correctly transferred from csv files into DataFrames. 
df_2017.head()
```

```python
#check that the 2016 data has correctly transferred from csv files into DataFrames. 
df_2017.shape
```

```python
#Merge the data from all the DataFrames based on the same "businessuei" data for every year. 
merged_df = df_2015.merge(df_2017, on='businessuei', how='outer')
```

```python
# Remove duplicates based on 'businessuei'
merged_df.drop_duplicates(subset='businessuei', keep='first', inplace=True)
```

```python
#select the columns to keep in the merged DataFrame
selected_columns = [
    'businessuei', '2015_total_dollars_obligated', '2015_total_outlayed_amount_for_overall_award', 
    '2015_current_total_value_of_award', '2015_base_and_all_options_value', '2015_potential_total_value_of_award', 
    '2015_number_of_offers_received', '2016_total_dollars_obligated', '2016_total_outlayed_amount_for_overall_award',
    '2016_current_total_value_of_award', '2016_base_and_all_options_value', '2016_potential_total_value_of_award',
    '2016_number_of_offers_received', '2017_total_dollars_obligated', '2017_total_outlayed_amount_for_overall_award',
    '2017_current_total_value_of_award', '2017_base_and_all_options_value', '2017_potential_total_value_of_award',
    '2017_number_of_offers_received','businessname', 'city', 'state', 'zipcode', 'naics', 'naicsdescription',
    'alaskan_native', 'american_indian', 'indian_tribe', 'native_hawaiian', 'tribally_owned', 'veteran_owned',
    'service_disabled_veteran', 'woman_owned', 'wosb', 'econ_disadv_wosb', 'jointventure_wosb', 
    'jointventure_econ_disadv_wosb', 'minority_owned', 'subcontinent_asian_indian_american', 'asian_pacific_american',
    'black_american', 'hispanic_american', 'native_american', 'other_minority'
]

merged_df = merged_df[selected_columns]
merged_df.head()
```

```python
#check the number of rows and columns in the newly merged DataFrame with all contracting data from 2015 to 2022
merged_df.shape
```

```python
unique_rows_based_on_multiple_columns = len(merged_df.drop_duplicates(subset=['businessuei', 'businessname']))
print("Number of unique rows based on multiple columns:", unique_rows_based_on_multiple_columns)
```

```python
#save the merged DataFrame to a new csv file in the same directory
merged_df.to_csv(directory_path + "All_2015to2017_Contracting.csv", index=False)
```
