## PPP Data Transformation and Cleanup:###

(a) delete the columns "ProcessingMethod", "CurrentApprovalAmount", and "UndisbursedAmount", 
(b) remove every "BusinessName" titled "NOT AVAILABLE", and every "BusinessAgeDescription" titled "New Business or 2 years or less" or "Startup, Loan Funds will Open Business", 
(c) in columns "HubzoneIndicator" and "LMIIndicator", replace Y with 1 and N with 0 to make the variables numerical and binary to facilitate the analysis (1 for Yes; 0 for No) 
(d) Save the file as PPP_up_to_150k_Aggregate.csv 

```python
import pandas as pd
import os
```

```python
#identify source folder:
source_folder = 'F:\\FAU PhD\\DISSERTATION\\DATABASES\\DataTESTS\\2023.08.27_Test'
source_file = 'ALL_PPP_up_to_150k_copy.csv'
destination_file = 'PPP_up_to_150k_Aggregate.csv'
```

```python
#read the csv file into a DataFrame
df = pd.read_csv(f"{source_folder}\\{source_file}")
df.head()
```

```python
#Delete the columns "ProcessingMethod", "CurrentApprovalAmount", and "UndisbursedAmount"
df.drop(columns=['ProcessingMethod', 'CurrentApprovalAmount', 'UndisbursedAmount'], inplace=True)

#Filter column "BusinessName" and remove every row for "NOT AVAILABLE"
df = df[df['BusinessName'] != 'NOT AVAILABLE']

#Filter column "BusinessAgeDescription" and remove specific rows
df = df[~df['BusinessAgeDescription'].isin(['New Business or 2 years or less', 'Startup, Loan Funds will Open Business'])]

#In columns "HubzoneIndicator" and "LMIIndicator", replace Y with 1 and N with 0
df['HubzoneIndicator'].replace({'Y': 1, 'N': 0}, inplace=True)
df['LMIIndicator'].replace({'Y': 1, 'N': 0}, inplace=True)

#preview dataFrame
df.head()
```

```python
# (e) Save the file as PPP_up_to_150k_Aggregate.csv
df.to_csv(f"{source_folder}\\{destination_file}", index=False)
```
