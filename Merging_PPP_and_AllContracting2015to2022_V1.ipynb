```python
#importing dependencies
import pandas as pd
```

```python
#define the directory path where each original file is located
directory_path = "F:\\FAU PhD\\DISSERTATION\\DATABASES\\DataTESTS\\2023.08.27_Test\\FinalDataMerge\\"
```

```python
#read each original csv file into a a Pandas DataFrame
df_PPP = pd.read_csv(directory_path + "PPP_up_to_150k_Aggregate.csv")
df_Contr = pd.read_csv(directory_path + "All_2015to2022_Contracting.csv")
```

```python
#merge the two DataFrames based on common business names, "businessname" column
merged_df = df_Contr.merge(df_PPP, on='businessname', how='outer')
```

```python
#Remove any duplicates entries from the merged DataFrame based on 'businessname'
merged_df.drop_duplicates(subset='businessname', keep='first', inplace=True)
```
